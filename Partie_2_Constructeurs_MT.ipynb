##### Import

```python
import numpy as np
import pandas as pd
```

##### Lecture des CSV

```python
df = pd.read_csv("Constructeurs_Clean_Mix.csv", sep=';')
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values(by = "date")
df = df.set_index("date")
df
```

##### Sélection des données entre 2 périodes (temporelles)

```python
mask = (df.index >= '2018-1-1') & (df.index <= '2018-12-31') 
df_bis = df.loc[mask]
print(len(df_bis))
```

##### Suppression de la colonne constructorRef qui fait doublon

```python
df_bis = df_bis.drop(columns = ["constructorRef"], axis=1)
df_bis
```

##### Sauvegarde du CSV

```python
df_bis.to_csv("df_test.csv", sep=';')
```
