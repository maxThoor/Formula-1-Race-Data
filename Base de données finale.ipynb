##### Import

```python
import pandas as pd
import numpy as np
```

##### Lecture des CSV

```python
df = pd.read_csv("Premier_Travail_Constructeurs_MT.csv", sep=';')
df
```

##### Suppression des colonnes qui font doublon ou qui ne nous intéressent pas

```python
df =df.drop(columns = ["year","laps"], axis=1)
```

##### Renommer les colonnes pour une meilleure compréhension

```python
df = df.rename(columns = {"date":"date_race","name":"constructor_name","nationality":"constructor_nationality","grid":"grid_position","positionOrder":"final_position","points":"points_scored","time_laps":"total_time_laps","rank":"rank_fastestLapTime"})
```

##### Ajout du nom des pilotes

```python
drivers = pd.read_csv("Datasets/drivers.csv", sep=',')
drivers = drivers.drop(columns = ["number","driverRef","url","dob"])
drivers = drivers.rename(columns = {'nationality':'driver_nationality','forename':'driver_forename','surname':'driver_surname'})

df = df.set_index("driverId")
drivers = drivers.set_index("driverId")

df_int = df.join(drivers)

df_int = df_int.reset_index()
```

##### Ajout des victoires des pilotes et donc des constructeurs

```python
df_int["win_driver"] = 0
df_int["win_constructor"] = 0

#for index, row in df_int.iterrows():
#    if df_int.loc[index,'final_position'] == 1:
for val in df_int["final_position"]:
    if val == 1:
        df_int.loc[df_int['final_position'] == 1, ['win_driver']] = val
        df_int.loc[df_int['final_position'] == 1, ['win_constructor']] = val
        
df_int
```

##### Ajout de la localisation du circuit

```python
circuits = pd.read_csv("Datasets/circuits.csv", sep=',')
circuits = circuits[["circuitId","location","country"]]
circuits = circuits.rename(columns = {"location":"location_circuit","country":"country_circuit"})

df_int = df_int.set_index("circuitId")
circuits = circuits.set_index("circuitId")

df_int2 = df_int.join(circuits)
df_int2 = df_int2.reset_index()
df_int2
```
