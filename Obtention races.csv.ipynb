```python
import pandas as pd
```

```python
circuits = pd.read_csv("Datasets_Kaggle/circuits.csv", sep=',')
races = pd.read_csv("Datasets_Kaggle/races.csv", sep=',')
results = pd.read_csv("Datasets_Kaggle/results.csv", sep=',')
status = pd.read_csv("Datasets_Kaggle/status.csv", sep=',')

print("circuits :", len(circuits))
print("races :", len(races))
print("results :", len(results))
```

##### Join circuits et races

```python
races = races.drop(columns = ["name","url","year"], axis=1)
circuits = circuits.drop(columns = ["url","lat","lng","alt"], axis=1)

circuits = circuits.set_index("circuitId")
races = races.set_index("circuitId")

mix1 = circuits.join(races)
mix1 = mix1.rename(columns = {'name':'circuitName','location':'circuitLocation','country':'circuitCountry'})
mix1
```

```python
results
```

##### Join Mix1 et results 

```python
mix1 = mix1.reset_index()
results = results.drop(columns = ["position","positionText","milliseconds"])
results = results.rename(columns = {'time':'laps_time','grid':'grid_position','positionOrder':'final_position','rank':'rank_fastestDriver'})


results = results.set_index("raceId")
mix1 = mix1.set_index("raceId")

mix2 = results.join(mix1)
mix2
```

##### Join Mix2 et status

```python
mix2 = mix2.reset_index()
status = status.set_index("statusId")
mix2 = mix2.set_index("statusId")

mix3 = mix2.join(status)
mix3

```

```python
print(len(mix3))
```

##### Ré arrangement de l'ordre des colonnes

```python
mix3 = mix3.reset_index()

classe = mix3[["date","circuitId","circuitName","circuitRef","circuitLocation","circuitCountry","round","raceId","time","resultId","driverId","constructorId","number","grid_position","final_position","points","laps","laps_time","fastestLap","fastestLapTime","rank_fastestDriver","fastestLapSpeed","statusId","status"]]

classe["date"] = pd.to_datetime(classe["date"])
classe = classe.sort_values(by = "date")
classe = classe.set_index("date")

classe
```

```python
classe.to_csv("races_Clean.csv", sep=';')
```
